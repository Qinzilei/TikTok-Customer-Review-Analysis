# TikTok Google Play Reviews — Exploratory Analysis

Load cleaned review data and explore volume, ratings, and sentiment distributions.

```python
import sys
from pathlib import Path

ROOT = Path.cwd().parent
sys.path.insert(0, str(ROOT))

import pandas as pd
import plotly.express as px

from config import PROCESSED_PARQUET
```

```python
df = pd.read_parquet(PROCESSED_PARQUET)
df["review_date"] = pd.to_datetime(df["review_date"])
print(f"Reviews: {len(df):,}  |  Avg rating: {df['score'].mean():.2f}")
df.head()
```

```python
daily = df.groupby("review_date").agg(
    count=("reviewId", "count"), avg_rating=("score", "mean")
).reset_index()
fig = px.bar(daily, x="review_date", y="count", title="Daily Review Volume")
fig.show()
```

```python
fig = px.pie(df, names="sentiment_label", title="VADER Sentiment Distribution")
fig.show()
```
